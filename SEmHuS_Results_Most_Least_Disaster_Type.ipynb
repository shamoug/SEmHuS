```
import pandas as pd
import pandas_profiling
from pandas_profiling import ProfileReport

import time
import re
import sys
import os
import numpy as np
import matplotlib.pyplot as plt
import nltk
import csv
import ast
import pickle
import itertools
import random
from scipy.spatial import distance
from ast import literal_eval
```

```
resultFI = './Outputs/RIVEC300_SMALL_Validated_Results_EM_DAT_Alpha_All.csv'
resultDF = pd.read_csv(resultFI)
```

```
list_alpha = resultDF['Alpha'].unique()

count_all = []

for a in list_alpha:
    alphaDF = resultDF.loc[(resultDF['Alpha'] == a)]
    
    list_tp, list_fp, list_tn, list_fn = [], [], [], []

    for idx, row in alphaDF.iterrows():
        list_tp.extend(literal_eval(row['True Positive']))
        list_fp.extend(literal_eval(row['False Positive']))
        list_tn.extend(literal_eval(row['True Negative']))
        list_fn.extend(literal_eval(row['False Negative']))

    count_Earthquake = [a, 'Earthquake', sum(map(lambda x : x == 'Earthquake', list_tp)), sum(map(lambda x : x == 'Earthquake', list_fp)), sum(map(lambda x : x == 'Earthquake', list_tn)), sum(map(lambda x : x == 'Earthquake', list_fn))]
    count_Volcano = [a, 'Volcano', sum(map(lambda x : x == 'Volcano', list_tp)), sum(map(lambda x : x == 'Volcano', list_fp)), sum(map(lambda x : x == 'Volcano', list_tn)), sum(map(lambda x : x == 'Volcano', list_fn))]
    count_Storm = [a, 'Storm', sum(map(lambda x : x == 'Storm', list_tp)), sum(map(lambda x : x == 'Storm', list_fp)), sum(map(lambda x : x == 'Storm', list_tn)), sum(map(lambda x : x == 'Storm', list_fn))]
    count_Temperature = [a, 'Extreme Temperature', sum(map(lambda x : x == 'Extreme Temperature', list_tp)), sum(map(lambda x : x == 'Extreme Temperature', list_fp)), sum(map(lambda x : x == 'Extreme Temperature', list_tn)), sum(map(lambda x : x == 'Extreme Temperature', list_fn))]
    count_Flood = [a, 'Flood', sum(map(lambda x : x == 'Flood', list_tp)), sum(map(lambda x : x == 'Flood', list_fp)), sum(map(lambda x : x == 'Flood', list_tn)), sum(map(lambda x : x == 'Flood', list_fn))]
    count_Landslide = [a, 'Landslide', sum(map(lambda x : x == 'Landslide', list_tp)), sum(map(lambda x : x == 'Landslide', list_fp)), sum(map(lambda x : x == 'Landslide', list_tn)), sum(map(lambda x : x == 'Landslide', list_fn))]
    count_Drought = [a, 'Drought', sum(map(lambda x : x == 'Drought', list_tp)), sum(map(lambda x : x == 'Drought', list_fp)), sum(map(lambda x : x == 'Drought', list_tn)), sum(map(lambda x : x == 'Drought', list_fn))]
    count_Fire = [a, 'Fire', sum(map(lambda x : x == 'Fire', list_tp)), sum(map(lambda x : x == 'Fire', list_fp)), sum(map(lambda x : x == 'Fire', list_tn)), sum(map(lambda x : x == 'Fire', list_fn))]
    count_Epidemic = [a, 'Epidemic', sum(map(lambda x : x == 'Epidemic', list_tp)), sum(map(lambda x : x == 'Epidemic', list_fp)), sum(map(lambda x : x == 'Epidemic', list_tn)), sum(map(lambda x : x == 'Epidemic', list_fn))]
    count_Insect = [a, 'Insect', sum(map(lambda x : x == 'Insect', list_tp)), sum(map(lambda x : x == 'Insect', list_fp)), sum(map(lambda x : x == 'Insect', list_tn)), sum(map(lambda x : x == 'Insect', list_fn))]

    count_all.extend([count_Earthquake, count_Volcano, count_Storm, count_Temperature, count_Flood, count_Landslide, count_Drought, count_Fire, count_Epidemic, count_Insect])
    # count_all.extend([count_Earthquake, count_Storm, count_Flood, count_Landslide, count_Drought, count_Epidemic])

print(count_all)
```

```
outputFI = './Outputs/RIVEC300_SMALL_Counted_Results_Summary_All.csv'
finalDF = pd.DataFrame(count_all, columns =['Alpha', 'Disaster', 'TP', 'FP', 'TN', 'FN'])

finalDF['Accuracy'] = (finalDF['TP']+finalDF['TN']) / (finalDF['TP']+finalDF['TN']+finalDF['FP']+finalDF['FN'])
finalDF['Precision'] = (finalDF['TP']) / (finalDF['TP']+finalDF['FP'])
finalDF['Recall'] = (finalDF['TP']) / (finalDF['TP']+finalDF['FN'])
finalDF['F1_Score'] = 2 * (finalDF['Recall'] * finalDF['Precision']) / (finalDF['Recall'] + finalDF['Precision'])

finalDF.to_csv(outputFI)
```
